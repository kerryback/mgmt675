```python
import pandas as pd
metrics = pd.read_excel("https://www.dropbox.com/scl/fi/pfy3zva24o878t81enp9w/metrics_5year.xlsx?rlkey=289wbumfl0q2b7d7k93iexm27&dl=1")
metrics.head()
```

```python
metrics.date = metrics.date.astype(str)
metrics.head()
```

```python
metrics.to_excel("metrics_5year.xlsx", index=False)
```

```python
years = [d.year for d in metrics.date]
filtr = [y in (2003, 2008, 2013, 2018, 2023) for y in years]
metrics2 = metrics[filtr]
```

```python
metrics2.to_excel("metrics_annual.xlsx")
```
