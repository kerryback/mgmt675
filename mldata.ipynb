```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set_style("whitegrid")

np.random.seed(0)
X = np.random.normal(scale = 2, size=(1000, 100))
errs = np.random.normal(size=1000)
```

```python
y1 = 3 * np.sin(2 * X[:,0]) + X[:,0]
y2 = np.where(y1-X[:,0]>2, 1, 0)
y3 = np.where(y1-X[:,0]>2, "A", "C")
y3 = np.where(y1-X[:,0]>-2, y3, "B")
y1 = pd.Series(y1)
y2 = pd.Series(y2)
y3 = pd.Series(y3)
df = pd.concat((y1, y2, y3, pd.DataFrame(X)), axis=1)
df.columns = ["continuous", "binary", "categorical"] + ["x" + str(i) for i in range(1, 101)]

fig, (ax2, ax3) = plt.subplots(1, 2)
ax2.scatter(df.x1, df.binary, alpha=0.2, s=2)
ax2.set_ylabel("binary target")
ax2.set_yticks([0, 1])
ax2.set_xlabel("feature 1")
ax3.scatter(df.x1, df.categorical, alpha=0.2, s=2)
ax3.set_ylabel("categorical target")
ax3.set_xlabel("feature 1")
ax3.set_yticks(["C", "B", "A"])
fig.tight_layout()
plt.savefig("data01.jpg")
```

```python
fig, ax1 = plt.subplots() 
ax1.scatter(df.x1, df.continuous, alpha=0.5, s=4)
ax1.set_ylabel("continuous target")
ax1.set_xlabel("feature 1")
plt.savefig("data00.jpg")
plt.show()
```

```python
y1 = 3 * np.sin(2 * X[:,0]) + X[:,0] + 2.5*errs
y2 = np.where(y1-X[:,0]>2, 1, 0)
y3 = np.where(y1-X[:,0]>2, "A", "C")
y3 = np.where(y1-X[:,0]>-2, y3, "B")
y1 = pd.Series(y1)
y2 = pd.Series(y2)
y3 = pd.Series(y3)
df = pd.concat((y1, y2, y3, pd.DataFrame(X)), axis=1)
df.columns = ["continuous", "binary", "categorical"] + ["x" + str(i) for i in range(1, 101)]
df.to_excel("mldata.xlsx")

fig, (ax2, ax3) = plt.subplots(1, 2)
ax2.scatter(df.x1, df.binary, alpha=0.2, s=2)
ax2.set_ylabel("binary target")
ax2.set_yticks([0, 1])
ax2.set_xlabel("feature 1")
ax3.scatter(df.x1, df.categorical, alpha=0.2, s=2)
ax3.set_ylabel("categorical target")
ax3.set_xlabel("feature 1")
ax3.set_yticks(["C", "B", "A"])
fig.tight_layout()
plt.savefig("data11.jpg")
```
